```python
#AUD Assignment 1
import pandas as pd
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import OneHotEncoder


#Reading the input dataset, which is a collection of reviews
data = pd.read_csv('AUD_Project_Data.csv')
d = pd.DataFrame(data)
print(d)
d

```

```python
#Merging Ad Title and Ad Description to Ad Details  

d['Ad Details'] = d['Ad Title'] + " " + d['Ad Description']

d
```

```python
#Converting Ad Details column to a list

doc_list = d['Ad Details'].values.tolist()
print(doc_list)
```

```python
#Tokenize each furniture ad detail 
tokenizer = nltk.tokenize.WhitespaceTokenizer() #used whitespace so that $ and price are together, hyperlink is together, etc
token_d_list = list()

for i in range(0,len(doc_list)):    
    Token_d = tokenizer.tokenize(doc_list[i])    
    token_d_list.append(Token_d)
print(token_d_list)
print(len(token_d_list))
```

```python
#Lemmatize words in each furniture ad detail and convert them to one case(lower)

lemmatizer = nltk.stem.WordNetLemmatizer()
lemmatized_token_d_list = list()
for j in range(0, len(token_d_list)):
    lemmatized_token_d = [lemmatizer.lemmatize(token).lower() for token in token_d_list[j]]
    lemmatized_token_d_list.append(lemmatized_token_d)
print(lemmatized_token_d_list)
print(len(lemmatized_token_d_list))
```

```python
#Remove stop-words from furniture details (Have retained numbers and punctuations to see prices and sizes of furnitures)
stop_words_removed_list = list()
for k in range(0, len(lemmatized_token_d_list)):
    stop_words_removed = [token for token in lemmatized_token_d_list[k] if not token in stopwords.words('english')]
    stop_words_removed_list.append(stop_words_removed)
    
print(stop_words_removed_list)
print(len(stop_words_removed_list))    
```

```python
#TF IDF Vectorization

tf_idf_reviews_gram = list()
for i in range(0, len(stop_words_removed_list)):
    tf_idf_reviews_gram.append(" ".join(stop_words_removed_list[i]))
                               
print(tf_idf_reviews_gram)  
```

```python
#Used N-gram of 1 and 2

vectorizer = TfidfVectorizer(ngram_range=(1,2), min_df=2) #will change min_df to 3 when we have more documents to test
vectorizer.fit(tf_idf_reviews_gram)
print(vectorizer.vocabulary_)
```

```python
TF_IDF_GRAM = vectorizer.transform(tf_idf_reviews_gram)
print(TF_IDF_GRAM.toarray())

```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.naive_bayes import GaussianNB
    
def simple_logistic_classify(X_tr, y_tr, X_test, y_test, description, _C=1.0):
    model = LogisticRegression(C=_C).fit(X_tr, y_tr)
    score = model.score(X_test, y_test)
    print('Test Score with', "Logistic Regression", score)
    return model

def SVM_classif(X_tr, y_tr, X_test, y_test):
    #Create a svm Classifier
    clf = svm.SVC(kernel='linear') # Linear Kernel
    #Train the model using the training sets
    clf.fit(X_tr, y_tr)
    #Predict the response for test dataset
    y_pred = clf.predict(X_test)
    score = clf.score(X_test, y_test)
    print('Test Score with',"SVM", score)
    
def Neural_Network(X_tr, y_tr, X_test, y_test):
    #Create a svm Classifier
    clf = MLPClassifier(hidden_layer_sizes=(150,150,150), random_state=1, max_iter=70)
    #Train the model using the training sets
    clf.fit(X_tr, y_tr)
    #Predict the response for test dataset
    y_pred = clf.predict(X_test)
    score = clf.score(X_test, y_test)
    print('Test Score with',"Neural", score)

def Random_Forest(X_tr, y_tr, X_test, y_test):
    #Create a svm Classifier
    clf = RandomForestClassifier(max_depth=55, random_state=0)
    #Train the model using the training sets
    clf.fit(X_tr, y_tr)
    #Predict the response for test dataset
    y_pred = clf.predict(X_test)
    score = clf.score(X_test, y_test)
    print('Test Score with',"RF", score)

def NaiveBayes(X_tr, y_tr, X_test, y_test):
    #Create a svm Classifier
    clf = GaussianNB()
    #Train the model using the training sets
    clf.fit(X_tr.toarray(), y_tr)
    #Predict the response for test dataset
    y_pred = clf.predict(X_test.toarray())
    score = clf.score(X_test.toarray(), y_test)
    print('Test Score with',"NaiveBayes", score)
```
