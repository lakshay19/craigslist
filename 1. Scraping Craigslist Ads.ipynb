```python
# importing the required selenium libraries
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
chrome_options = Options()
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver', options = chrome_options)
```

```python
# url for the first quotes page
url = "https://chicago.craigslist.org/d/furniture/search/fua"

# opening the url in selenium
driver.get(url)
```

```python
# creating empty lists for storing the information
links = []

# looping through all the pages
for i in range(1,11):
    
    html = driver.page_source # Get the html of the page
    
    # parsing url through soup
    soup = bs(html, 'html.parser')
        
    # louping through each quote
    for link in soup.find_all('a', class_ = "result-image gallery"): 
        links.append(link['href'])   

    time.sleep(np.random.uniform(1,5))
    
    # next page
    driver.get(url + "?s=" + str(120*i))
```

```python
links
```

```python
print(len(links))
```

```python
text = ""
```

```python
def find_text(base):
    global text

    if str(type(base.next_sibling)) != "<class 'bs4.element.Tag'>":
        text += base.next_sibling

    if base.next_sibling.next_sibling:
        find_text(base.next_sibling)
```

```python
ad_title = []
ad_description = []


for i in range(0,len(links)):

    time.sleep(np.random.uniform(1,5))

    driver.get(links[i])
    
    html = driver.page_source # Get the html of the page

    soup = bs(html, 'html.parser')
    
    ad_title.append(soup.find('span', id = "titletextonly").string)

    find_text(soup.find('div', class_ = "print-information print-qrcode-container"))

    ad_description.append(text)
    
    text = ""
```
